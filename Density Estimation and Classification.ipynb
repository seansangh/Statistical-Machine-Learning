```python
import numpy as np
import pandas as pd

myID='6771' #change to last 4 digit of your studentID
geneNewData.geneData(myID)
Numpyfile0 = scipy.io.loadmat('digit0_stu_train'+myID+'.mat')
Numpyfile1 = scipy.io.loadmat('digit1_stu_train'+myID+'.mat')
Numpyfile2 = scipy.io.loadmat('digit0_testset'+'.mat')
Numpyfile3 = scipy.io.loadmat('digit1_testset'+'.mat')
train0 = Numpyfile0.get('target_img')
train1 = Numpyfile1.get('target_img')
test0 = Numpyfile2.get('target_img')
test1 = Numpyfile3.get('target_img')
print([len(train0),len(train1),len(test0),len(test1)])
print('\r')



print(len(train0[0][0]))
print(numpy.mean(train0[0]))
print(numpy.std(train0[0]))

training0_feat1 = []
training0_feat2 = []

for image in train0:
    training0_feat1.append(numpy.mean(image))
    training0_feat2.append(numpy.std(image))
print(len(training0_feat1))
print(len(training0_feat2))

training1_feat1 = []
training1_feat2 = []

for image in train1:
    training1_feat1.append(numpy.mean(image))
    training1_feat2.append(numpy.std(image))
print(len(training1_feat1))
print(len(training1_feat2))

#Mean of feature1 for digit0
meanf1d0 = numpy.mean(training0_feat1)

#Variance of feature1 for digit0
varf1d0 = numpy.var(training0_feat1)

#Mean of feature2 for digit0
meanf2d0 = numpy.mean(training0_feat2)

#Variance of feature2 for digit0
varf2d0 = numpy.var(training0_feat2)

#Mean of feature1 for digit1
meanf1d1 = numpy.mean(training1_feat1)

#Variance of feature1 for digit1
varf1d1 = numpy.var(training1_feat1)

#Mean of feature2 for digit1
meanf2d1 = numpy.mean(training1_feat2)

#Variance of feature2 for digit1
varf2d1 = numpy.var(training1_feat2)


def prob(value, mean, var):
    return (1 / numpy.sqrt(2*numpy.pi*var)) * numpy.exp(-numpy.square(value - mean)/(2*var))

results = []
for image in test0:
    image_mean = numpy.mean(image)
    image_std = numpy.std(image)
    prob0 = 0.5
    prob1 = 0.5
    prob0 *= (prob(image_mean, meanf1d0, varf1d0) * prob(image_std, meanf2d0, varf2d0))
    prob1 *= (prob(image_mean, meanf1d1, varf1d1) * prob(image_std, meanf2d1, varf2d1))
    results.append([prob0,prob1])
pred0 = [0,0]
for test in results:
    pred0[numpy.argmax(test)] += 1
    
print(pred0)


results = []
for image in test1:
    image_mean = numpy.mean(image)
    image_std = numpy.std(image)
    prob0 = 0.5
    prob1 = 0.5
    prob0 *= (prob(image_mean, meanf1d0, varf1d0) * prob(image_std, meanf2d0, varf2d0))
    prob1 *= (prob(image_mean, meanf1d1, varf1d1) * prob(image_std, meanf2d1, varf2d1))
    results.append([prob0,prob1])
pred1 = [0,0]
for test in results:
    pred1[numpy.argmax(test)] += 1
    
print(pred1)


#Accuracy
# The # correct / total for that class
print(f"Test0's accuracy: {pred0[0] / numpy.sum(pred0)}")
print(f"Test1's accuracy: {pred1[1] / numpy.sum(pred1)}")
```

```python
myID='6771' #change to last 4 digit of your studentID
geneNewData.geneData(myID)
Numpyfile0 = scipy.io.loadmat('digit0_stu_train'+myID+'.mat')
Numpyfile1 = scipy.io.loadmat('digit1_stu_train'+myID+'.mat')
Numpyfile2 = scipy.io.loadmat('digit0_testset'+'.mat')
Numpyfile3 = scipy.io.loadmat('digit1_testset'+'.mat')
train0 = Numpyfile0.get('target_img')
train1 = Numpyfile1.get('target_img')
test0 = Numpyfile2.get('target_img')
test1 = Numpyfile3.get('target_img')
print([len(train0),len(train1),len(test0),len(test1)])
print('\r')



print(len(train0[0][0]))
print(numpy.mean(train0[0]))
print(numpy.std(train0[0]))

#TASK 1: Extracting features from trainset

t0f1 = []  #train0 feature 1
t0f2 = []  #train0 feature 2

for img in train0:
    t0f1.append(numpy.mean(img))
    t0f2.append(numpy.std(img))

t1f1 = []  #train1 feature 1
t1f2 = []  #train1 feature 2

for img in train1:
    t1f1.append(numpy.mean(img))
    t1f2.append(numpy.std(img))


#TASK 2: Calculating Parameters for Naive Bayes Classifiers

mf1d0 = numpy.mean(t0f1)    #Mean of digit0 Feature 1
vf1d0 = numpy.var(t0f1)     #Variance of digit0 Feature 1
mf2d0 = numpy.mean(t0f2)    #Mean of digit0 Feature 2
vf2d0 = numpy.var(t0f2)     #Variance of digit0 Feature 2
mf1d1 = numpy.mean(t1f1)    #Mean of digit1 Feature 1
vf1d1 = numpy.var(t1f1)     #Variance of digit1 Feature 1
mf2d1 = numpy.mean(t1f2)    #Mean of digit1 Feature 2
vf2d1 = numpy.var(t1f2)     #Variance of digit1 Feature 2

#print(mf1d0,vf1d0,mf2d0,vf2d0,mf1d1,vf1d1,mf2d1,vf2d1, sep="; ")


#TASK 3: Implementing Calculation Formula To Classify/Predict Labels

myList=[]
myList2=[]
def myFunc(value, mean, var):
    return (1 / numpy.sqrt(2*numpy.pi*var)) * numpy.exp(-numpy.square(value - mean)/(2*var))

for images in test0:
    t0mean = numpy.mean(images)
    t0std = numpy.std(images)
    prob0 = 0.5
    prob1 = 0.5
    prob0 *= (myFunc(t0mean, mf1d0, vf1d0) * myFunc(t0std, mf2d0, vf2d0))
    prob1 *= (myFunc(t0mean, mf1d1, vf1d1) * myFunc(t0std, mf2d1, vf2d1))
    myList.append([prob0,prob1])

for images in test1:
    t1mean = numpy.mean(images)
    t1std = numpy.std(images)
    prob0 = 0.5
    prob1 = 0.5
    prob0 *= (myFunc(t1mean, mf1d0, vf1d0) * myFunc(t1std, mf2d0, vf2d0))
    prob1 *= (myFunc(t1mean, mf1d1, vf1d1) * myFunc(t1std, mf2d1, vf2d1))
    myList2.append([prob0,prob1])

predict0 = [0,0]
predict1 = [0,0]

for test in myList:
    predict0[numpy.argmax(test)] += 1
#print(predict0)

for test in myList2:
    predict1[numpy.argmax(test)] += 1
#print(predict1)


#TASK 4: Accuracy Calculation

accuracy0= predict0[0] / numpy.sum(predict0)
accuracy1= predict1[1] / numpy.sum(predict1)

print("The Accuracy for digit0 is: ", accuracy0)
print("The Accuracy for digit1 is: ", accuracy1)


answers=[]
answers=[myID, mf1d0, vf1d0, mf2d0, vf2d0, mf1d1, vf1d1, mf2d1, vf2d1, accuracy0, accuracy1]
print(answers)
#['6771', 44.283860714285716, 114.41087554438542, 87.51159136983738, 100.35618061196351, 19.335694897959186, 31.346498038982716, 61.296561874734856, 82.04622496799617, 0.9173469387755102, 0.9233480176211454]
```
